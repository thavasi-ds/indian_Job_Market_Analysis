##**Which cities pay the most for Data Scientists**?

```python
df['salary_max_lpa'].value_counts()
```

```python
df['salary_midpoint_lpa'].value_counts()
```

```python
df_need1 = df[['job_title','primary_city','salary_max_lpa','salary_midpoint_lpa','salary_min_lpa']]
df_need1
```

```python
ds_df = df_need1 [df_need1['job_title'] == 'Data Scientist']
ds_df
```

```python
max_pay =ds_df.groupby(["primary_city","job_title"])[["salary_max_lpa"]].max()
max_pay.sort_values("salary_max_lpa",ascending=False).reset_index()
```
